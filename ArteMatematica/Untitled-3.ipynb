```python

import PyPDF2


import pandas as pd

import re

```

```python
#Read content of pdf
data = []



with open("c:\\Pessoal\\Vestibular 1998 - Candidatos convocados.pdf", "rb") as file:
    reader = PyPDF2.PdfReader(file)
    text = ""
    # Loop through all pages
    for page in reader.pages:

        text = page.extract_text() + "\n"
                
        if text:  # check if page has text
            for line in text.split("\n"):   # split into lines

                data.append({"page": len(data) + 1, "content": line})

df = pd.DataFrame(data)
print(df.head())

```

```python
#Procurando o conteúdo específico
#print(df.shape)
print(df['content'][2])
#print(df['content'][15])
#print(df['content'][16])
#print(df['content'][24])
#print(df['content'][25])

#print(df['content'][33])
#print(df['content'][34])

```

```python
linhas = [2, 15, 16, 24, 25, 33, 34]

dados =[]

for i in linhas:
    str1 = df['content'][i].replace("CANDIDATOS CONVOCADOS EM TODAS AS CHAMADASMINISTÉRIO DA DEFESA", " ")

    blocks = re.findall(r'[A-ZÀ-ÖØ-Ý ]+\d+(?:[ ,\d]+)?', str1)

    dados.append(blocks)

print(dados)
```

```python
print(len(dados))
```

```python
dados_2 = dados[0] + dados[1] + dados[2] + dados[3] + dados[4] + dados[5] + dados[6]
print(len(dados_2))
```

```python
new_dados=[]
for dado in dados_2:
    
    match = re.search(r'[A-ZÀ-ÖØ-Ý ]+(?=\s+\d)', dado)
    if match:
        name = match.group().strip()
        numbers = dado[len(name):].strip()
        new_dados.append((name, numbers))

print(new_dados)
print(len(new_dados))
```

```python
trat_dados =[]

for dado in new_dados:
    name = dado[0].title().strip()
    numbers = dado[1].replace(",", ".").split(' ')
    trat_dados.append((name, int(numbers[6])))
```

```python
df_dados = pd.DataFrame(trat_dados)
df_dados.head()
```
